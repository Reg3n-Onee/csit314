# Class Creation

```python
class Data_management:
    def __init__(self, stationID, componentID, reading, date_read, location, component_name, wind_direction):
        self._stationID = stationID
        self._componentID = componentID
        self._reading = reading
        self._date_read = date_read
        self._location = location
        self._component_name = component_name
        self._wind_direction = wind_direction
        self._reading = reading
    def get_stationID(self):
        return self._stationID
    def get_componentID(self):
        return self._componentID
    def get_reading(self):
        return self._reading
    def get_date_read(self):
        return self._date_read
    def get_location(self):
        return self._location
    def get_component_name(self):
        return self._component_name
    def get_wind_direction(self):
        return self._wind_direction
    
    
class maintenance:
    def __init__(self, componentID, component_name, reading, date_read, last_maintenance, wind_direction, status):
        self._componentID = componentID
        self._component_name = component_name
        self._reading = reading
        self._date_read = date_read
        self._last_maintenance = last_maintenance
        self._wind_direction = wind_direction
        self._status = status
    def get_componentID(self):
        return self._componentID
    def get_component_name(self):
        return self._component_name
    def get_reading(self):
        return self._reading
    def get_date_read(self):
        return self._date_read
    def get_last_maintenance(self):
        return self._last_maintenance
    def get_wind(self):
        return self._wind_direction
    def get_status(self):
        return self._status 

    
```
